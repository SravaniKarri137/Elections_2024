```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data into a DataFrame
# Adjust the path if needed based on where you uploaded the file
data = pd.read_csv('/content/lok_sabha_results.csv')

# Drop any rows with missing values (if any)
data.dropna(inplace=True)

# Sort the data by 'Won' column in descending order
data.sort_values(by='Won', ascending=False, inplace=True)

# Extracting top 10 parties by seats won
top_10_parties = data.head(10)

# Generate a bar plot of top parties by seats won (before choosing insight)
plt.figure(figsize=(10, 6))
plt.bar(top_10_parties['Party'], top_10_parties['Won'], color='skyblue')
plt.xlabel('Parties')
plt.ylabel('Seats Won')
plt.title('Top Parties by Seats Won')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Define all key insights
key_insights = [
    f"1. Dominance of BJP: BJP has {top_10_parties.loc[top_10_parties['Party'].str.contains('BJP'), 'Won'].values[0]} seats.",
    f"2. Top Five Parties: {', '.join(top_10_parties['Party'].head(5))} are prominent parties.",
    f"3. Regional Party Presence: SP, AITC, and DMK have substantial seats.",
    f"4. Single Seat Parties: {', '.join(top_10_parties[top_10_parties['Won'] == 1]['Party'])} have minimal representation.",
    f"5. Performance of New Parties: {', '.join(top_10_parties[top_10_parties['Won'] == 1]['Party'])} have emerged with minimal seats.",
    f"6. Independent Candidates: Independent candidates total {top_10_parties[top_10_parties['Party'] == 'Independent']['Won'].sum()}.",
    f"7. Party Fragmentation: The top 10 parties collectively hold {top_10_parties['Won'].sum()} out of {data['Won'].sum()} total seats.",
    f"8. Regional Parties Impact: Regional parties collectively won {data[data['Party'].str.contains('SP|AITC|DMK|TDP|JD')]['Won'].sum()} seats.",
    f"9. Minority Party Influence: Parties with less than 5 seats collectively account for {data[data['Won'] < 5]['Won'].sum()} seats.",
    f"10. Electoral Diversity: Over {data['Party'].nunique()} parties contested in the elections."
]

# Print all key insights for user to choose from
print("Choose an insight to display:")
for i, insight in enumerate(key_insights, 1):
    print(f"{i}. {insight}")

# Prompt user to choose an insight (by index)
while True:
    try:
        choice = int(input("Enter the number of the insight you want to display (or 0 to exit): "))
        if choice == 0:
            print("Exiting program.")
            break
        elif 1 <= choice <= len(key_insights):
            chosen_insight = key_insights[choice - 1]
            print("\nChosen Insight:")
            print(chosen_insight)

        else:
            print("Invalid choice. Please enter a number within the range or 0 to exit.")

    except ValueError:
        print("Invalid input. Please enter a number.")
```
